```python
df = pd.read_csv(r'AccidentesAviones.csv')
```

```python
#analisis columnas
'''
'Unnamed: 0': numeracion de los accidentes, indice
'fecha': fecha del accidente
'HORA declarada': hora accidente en formato 24hs
'Ruta': ubicacion del accidente
'OperadOR': aerolinea u operador de la aeronave
'flight_no': numero de vuelo asignado por operador
'route': ruta aerea completa o parcialmente transcurrida previa al accidente
'ac_type': tipo de aeronave
'registration':  registro ICAO de la aeronave
'cn_ln': numero de serie/construccion
'all_aboard': total de gente a bordo
'PASAJEROS A BORDO': pasajeros a bordo
'crew_aboard': tripulacion a bordo
'cantidad de fallecidos': cantidad de fallecidos
'passenger_fatalities':pasajeros fallecidos
'crew_fatalities': tripulantes fallecidos
'ground': fallecidos terrestres
'summary' breve descripcion del accidente y causa
'''
df.columns
```

```python
df.isnull().sum()       # 0
df.replace('?', np.NaN, inplace=True)
df.isnull().sum()
```

```python
# separamos la informacion sobre fatalities por si llegamos a necesitarla luego
df_fatalities = df[['Unnamed: 0','all_aboard', 'PASAJEROS A BORDO', 'crew_aboard', 'cantidad de fallecidos', 'passenger_fatalities', 'crew_fatalities', 'ground']]
df_fatalities.columns = ['id','total_a_bordo', 'pasajero_a_bordo', 'tripulantes_a_bordo', 'total_fallecidos', 'pasajeros_fallecidos', 'tripulantes_fallecidos', 'fallecidos_en_tierra']
```

```python
df_fatalities['total_a_bordo'] = df_fatalities['total_a_bordo'].fillna(1)
df_fatalities = df_fatalities.fillna(0)
```

```python
df_fatalities.isnull().sum()
```

```python
df_fatalities = df_fatalities.astype(int)
df_fatalities.id = df_fatalities.id.astype(str)
df_fatalities.info()
```

```python
# nos quedamos con las columnas que consideramos utiles para nuestro analisis
df = df[['Unnamed: 0','fecha','Ruta', 'OperadOR', 'route', 'ac_type']]
```

```python
#renombramos
df.columns = ['id','fecha', 'ubicacion', 'operador', 'ruta', 'tipo_aeronave']
df['fecha'] = pd.to_datetime(df['fecha'])
df.id = df.id.astype(str)
df = df.fillna('Sin Dato')
df.info()
```

```python
#df_fatalities.to_csv('fatalities.csv', index=False)
df.to_csv('accidents.csv', index=False)
```
